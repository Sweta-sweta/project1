```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
```

```python

data1=pd.read_csv("covid_19_clean_complete.csv")
data3=pd.read_csv("complete.csv")
print(data3.head())
print(data1.head())
```

In below code With the use of data3 i am going to group the data according to date

```python
Datewise=data3.groupby(['Date']).sum()
print(Datewise.head())
```

# DateWise Comparison among Cases,Death,Recovery in india 

```python
plt.figure(figsize=(20,10))
dates=Datewise.index.values
plt.xlabel('Date')
plt.ylabel('Affected Population')
plt.title('DateWise comparision among Cases, Death, Recovery in India in last 4 months')
plt.plot(dates,Datewise[["Death"]],label='Death')
plt.plot(dates,Datewise[["Cured/Discharged/Migrated"]],label='Recovered')
plt.plot(dates,Datewise[["Total Confirmed cases"]],label='Affected')
plt.plot(dates,Datewise[["Total Confirmed cases"]],color='Green',linewidth='3',linestyle="dashed",marker="o",markerfacecolor="green",markersize='7')
plt.plot(dates,Datewise[["Death"]],color='Blue',linewidth='3',linestyle="dashed",marker="o",markerfacecolor="Blue",markersize='7')
plt.plot(dates,Datewise[["Cured/Discharged/Migrated"]],color='Orange',linewidth='3',linestyle="dashed",marker="o",markerfacecolor="Orange",markersize='7')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

in above graph we can see that how the confirmed,Recovery and Death cases are increasing  . Before 10 march cases are around 0 but after 10th march it increased till now.

# state wise comparison

```python
StateWise=data3.groupby(["Name of State / UT"]).sum()
print(StateWise.head())
```

```python
plt.figure(figsize=(20,10))
State=StateWise.index.values
plt.bar(State,StateWise["Total Confirmed cases"],color='green',width=0.8)
plt.bar(State,StateWise["Death"],color='red',width=0.8)
plt.bar(State,StateWise["Cured/Discharged/Migrated"],color='Orange',width=0.8)
plt.xlabel("Affected states")
plt.ylabel("Affected case")
red_patch=mpatches.Patch(color="green",label="Cases")
yellow_patch=mpatches.Patch(color="red",label="Death")
green_patch=mpatches.Patch(color="orange",label="Recovery")
plt.legend(handles=[red_patch,yellow_patch,green_patch])
plt.title("Affected States in india")
plt.xticks(rotation=90)
```

In this above graph we can see that maharshtra has heighest no. of cases. Deaths are very less in all states thats why it is not visible in the graph .

```python
plt.figure(figsize=(20,10))
plt.xlabel("sates")
plt.ylabel("no. of people")
plt.title("Death in india")
plt.bar(State,StateWise["Death"],color='Red',width=0.8)
plt.xticks(rotation=90)
plt.show()
```

In the above graph we can see that the no. of death cases in india.we can see that 'Maharshtra' has heightest no. of death,'Madhya pradesh' is in second no and 'Gujarat' is in third place of death in india.

# Country wise data

```python
plt.figure(figsize=(20,10))
CountryWise=data1.groupby(["Country/Region"]).sum()
print(CountryWise.head())
Country=CountryWise.index.values
plt.xlabel("Countries")
plt.ylabel("Population affected")
plt.title("Affected countries in World-Wide")
plt.bar(Country,CountryWise["Confirmed"],color="Yellow",width=0.8)
plt.bar(Country,CountryWise["Deaths"],color="Red",width=0.8)
plt.bar(Country,CountryWise["Recovered"],color="Green",width=0.8)
red_patch=mpatches.Patch(color="green",label="recovery")
yellow_patch=mpatches.Patch(color="red",label="Death")
green_patch=mpatches.Patch(color="yellow",label="confirmed")
plt.legend(handles=[red_patch,yellow_patch,green_patch])
plt.xticks(rotation=90)
plt.show()
```

in the above graph we can see that US has heighst no. of confirmed cases but recovery is so less and china is in the second place of confirmed cases but recovery rate is better than US and here we can see that death rate is very less in all the countries thats why it is not visible in all countries data

```python
CountryWise["Deaths"].plot(kind='pie',figsize=(8, 9),autopct='%1.1f%%',startangle=90,shadow=False,)
plt.title('Population Affected by Total Percentage by Death')
plt.axis('equal') 
plt.show()
CountryWise["Confirmed"].plot(kind='pie',figsize=(8, 9),autopct='%1.1f%%',startangle=90,shadow=False,)
plt.title('Population Affected by Total Percentage by Confirmed')
plt.axis('equal') 
plt.show()
CountryWise["Recovered"].plot(kind='pie',figsize=(8, 9),autopct='%1.1f%%', startangle=90, shadow=False,)
plt.title('Population Affected by Total Percentage by Recovered')
plt.axis('equal') 

plt.show()
```

In above piechart one is for showing the death rate, 
                  2nd is for  showing confirmed cases, 
                  3rd is for recovery rate

In first piechart, we can see that heighest death rate is in Italy which is 20.6%, 2nd heighest is US with 15.5%,  3rd heighest is Spain with 15.0%

In Second piechart, we can see that the confirmed cases in throughtout the world . The heighest no. of confirmed cases is in US with 23.8% of total confirmed cases, 2nd heighest is China with 15.0% of total confirmed cases, Italy is 3rd heighest covid affected country.here confirmed population is 9.8% of the total confirmed cases

In third piechart, we can see that total recovery has been done in china is 40.6% which is heighest recovered country,2nd place of recovery is Spain and third is germany
